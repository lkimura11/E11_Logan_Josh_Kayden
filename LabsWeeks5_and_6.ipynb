```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Inside

```python
inside_data = pd.read_csv('inside_data.csv')
```

```python
fig1, ax1 = plt.subplots(figsize=(25,10),ncols=3,nrows=2)

ax1[0,0].plot(inside_data['Temperature'])
ax1[0,0].set_ylabel('Temperature (Celcius)')
ax1[0,0].set_xlabel('Time (Seconds)')
ax1[0,0].set_title('Temperature vs. Time Inside')

ax1[0,1].plot(inside_data['Gas'])
ax1[0,1].set_ylabel('Gas (Ohm)')
ax1[0,1].set_xlabel('Time (Seconds)')
ax1[0,1].set_title('Gas vs. Time Inside')

ax1[0,2].plot(inside_data['Relative Humidity'])
ax1[0,2].set_ylabel('Relative Humidity (%)')
ax1[0,2].set_xlabel('Time (Seconds)')
ax1[0,2].set_title('Relative Humidity vs. Time Inside')

ax1[1,0].plot(inside_data['Pressure'])
ax1[1,0].set_ylabel('Pressure (hPa)')
ax1[1,0].set_xlabel('Time (Seconds)')
ax1[1,0].set_title('Pressure vs. Time Inside')

ax1[1,1].plot(inside_data['Altitude'])
ax1[1,1].set_ylabel('Altitude (Meters)')
ax1[1,1].set_xlabel('Time (Seconds)')
ax1[1,1].set_title('Altitude vs. Time Inside')

ax1[1,2].plot(inside_data['PM 2.5 Standard'])
ax1[1,2].set_ylabel('PM 2.5 (CU)')
ax1[1,2].set_xlabel('Time (Seconds)')
ax1[1,2].set_title('PM 2.5 vs. Time Inside')
```

```python
fig1, ax1 = plt.subplots(figsize=(25,10),ncols=3,nrows=2)

ax1[0,0].hist(inside_data['Temperature'], bins = 17)
ax1[0,0].set_ylabel('Frequency')
ax1[0,0].set_xlabel('Temperature (Celcius)')
ax1[0,0].set_title('Temperature Frequency Inside')

ax1[0,1].hist(inside_data['Gas'], bins = 17)
ax1[0,1].set_ylabel('Frequency')
ax1[0,1].set_xlabel('Gas (Ohm)')
ax1[0,1].set_title('Gas Frequency Inside')

ax1[0,2].hist(inside_data['Relative Humidity'], bins = 17)
ax1[0,2].set_ylabel('Frequency')
ax1[0,2].set_xlabel('Relative Humidity (%)')
ax1[0,2].set_title('Relative Humidity Frequency Inside')

ax1[1,0].hist(inside_data['Pressure'], bins = 17)
ax1[1,0].set_ylabel('Frequency')
ax1[1,0].set_xlabel('Pressure (hPa)')
ax1[1,0].set_title('Pressure Frequency Inside')

ax1[1,1].hist(inside_data['Altitude'], bins = 17)
ax1[1,1].set_ylabel('Frequency')
ax1[1,1].set_xlabel('Altitude (Meters)')
ax1[1,1].set_title('Altitude Frequency Inside')

ax1[1,2].hist(inside_data['PM 2.5 Standard'], bins = 17)
ax1[1,2].set_ylabel('Frequency')
ax1[1,2].set_xlabel('PM 2.5 (CU)')
ax1[1,2].set_title('PM 2.5 Frequency Inside')
```

```python
for col in ['Temperature','Gas','Relative Humidity','Pressure','Altitude','PM 2.5 Standard']:
    print(f"Mean of {col} Inside: {inside_data[col].mean()}")
```

```python
for col in ['Temperature','Gas','Relative Humidity','Pressure','Altitude','PM 2.5 Standard']:
    print(f"STD of {col} Inside: {inside_data[col].std()}")
```

```python
for col in ['Temperature','Gas','Relative Humidity','Pressure','Altitude','PM 2.5 Standard']:
    print(f"Mean Uncerstainty of {col} Inside: {inside_data[col].std()/(np.sqrt(300))}")
```

# Correlation in Data

```python
fig1, ax1 = plt.subplots(figsize=(15,5),ncols=2)

ax1[0].plot(inside_data['Temperature'], inside_data['Relative Humidity'])
ax1[0].set_title('Temperature vs Relative Humidity Inside')
ax1[0].set_xlabel('Temperature (Celcius)')
ax1[0].set_ylabel('Relative Humidity (%)')

ax1[1].plot(cropped_outside['Temperature'], cropped_outside['Relative Humidity'])
ax1[1].set_title('Temperature vs Relative Humidity Outside')
ax1[1].set_xlabel('Temperature (Celcius)')
ax1[1].set_ylabel('Relative Humidity (%)')
```

The important variable when making these plots is time, and the times of the relative humidity and tempertature match up. Plot 1 shows a negative correleation between temperature and relative humidity while indoors. Plot 2 shows a slightly negative correlation between temperature and relative humidity outside, but the graph is very noisy. Because of this, it appears to be mostly uncorrelated outside. 

```python
fig1, ax1 = plt.subplots(figsize=(20,5),ncols=3)

ax1[0].plot(inside_data['Temperature'], inside_data['Gas'])
ax1[0].set_title('Temperature vs Gas Inside')
ax1[0].set_xlabel('Temperature (Celcius)')
ax1[0].set_ylabel('Gas (Ohm)')

ax1[1].plot(inside_data['Temperature'], inside_data['Pressure'])
ax1[1].set_title('Temperature vs Pressure Inside')
ax1[1].set_xlabel('Temperature (Celcius)')
ax1[1].set_ylabel('Pressure (hPA)')

ax1[2].plot(inside_data['Temperature'], inside_data['Altitude'])
ax1[2].set_title('Temperature vs Relative Humidity Inside')
ax1[2].set_xlabel('Temperature (Celcius)')
ax1[2].set_ylabel('Altitude (Meters)')
```

For inside, there is a clear positive correlation between temperature and gas but not for other data types.
