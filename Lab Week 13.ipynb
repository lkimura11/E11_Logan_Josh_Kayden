```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('LabWeek13Locations.csv')
data = data.iloc[35:].reset_index(drop=True)
data
```

## a.

```python
plt.figure(figsize=(20,10))


time_seconds = np.arange(len(data)) * 5
plt.plot(time_seconds, data['counts'])
plt.ylabel('Counts')
plt.xlabel('Time (seconds)')
plt.title('Counts vs. Time')
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import ScalarFormatter

fig = plt.figure(figsize=(20,7))
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(data['Longitude'], data['Latitude'], data['Altitude'], c=data['counts'], cmap='viridis')
cbar = plt.colorbar(sc, label='Radiation Count')
cbar.ax.set_ylabel('Radiation Count', labelpad=15) 

formatter = ScalarFormatter()
formatter.set_scientific(False)

ax.xaxis.set_major_formatter(formatter)
ax.yaxis.set_major_formatter(formatter)
ax.zaxis.set_major_formatter(formatter)

for tick in ax.get_xticklabels():
    tick.set_rotation(30)
    tick.set_fontsize(8)

for tick in ax.get_yticklabels():
    tick.set_rotation(15)
    tick.set_fontsize(8)

ax.set_xlabel('Longitude', labelpad=20)
ax.set_ylabel('Latitude', labelpad=10)
ax.set_zlabel('Altitude (m)')
ax.set_title('Counts by Location')
plt.show()
```

```python
print(f"Counts: mean:{np.mean(data['counts']):.3f}, std: {np.std(data['counts']):.3f}")
print(f"PM2.5: mean:{np.mean(data['PM25']):.3f}, std: {np.std(data['PM25']):.3f}")
print(f"Temperature: mean:{np.mean(data['Temperature']):.3f}, std: {np.std(data['Temperature']):.3f}")
```

The counts and temperature data look like they follow a normal distribution, but the PM 2.5 data looks like it does not. However, our PM 2.5 data is likely scewed because we were to a location for a short period of time where there was lots of vapor, which explains why it does not follow a normal distribution. 

## c.

```python
N = np.mean(data['counts'])
std_counting = np.sqrt(N)

print(f"Counting statistics variance estiamte: {N:.3f}")
print(f"Counting statistics STD estimate: {std_counting:.3f}")
print(f"Measured STD: {np.std(data['counts']):.3f}")
```

## d.

```python
N = np.mean(data['counts'])
N_samples = 151
unc_measured_std = (np.sqrt(N))/(np.sqrt(2*np.sqrt(N_samples)))
print(f"Uncertainty on measured STD from CLT: {unc_measured_std:.3f}")
```

## e

The standard deviation estimate for the data set based on counting statistics is 5.651, while the measured standrad deviation is 5.273. The difference is 0.378, which is less than 1.140, meaning there is less variation in the data than we would expect from simple counting statistics. Since we do not see more variation than expected, this implies that the sources of radiation contributing to our data, most likely background radiation, is statistically uniform and follows a Poisson distribution.
